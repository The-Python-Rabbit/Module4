```python

class Employee():
    def __init__(self, empid, name, address, date, salary):
        self.__empid = int(empid)
        self.__name = name
        self.__address = address
        self.__date = date
        self.__salary = int(salary)
        
    def print_emp_info(self):
        return (f"\nEmployee ID: {self.__empid}\tName: {self.__name}\nAddress: {self.__address}\t\tStart Date: {self.__date}\tSalary: {self.__salary}")

#
# Use this method for writing information to file in same format as input file.
    def print_emp_info_brief(self):
        return(f"{self.__empid},{self.__name},{self.__address},{self.__date},{self.__salary}")
    
    def update_employee_name(self, name):
        self.__name = name
        
    def update_employee_address(self, address):
        self.__address = address
        
    def update_employee_salary(self, salary):
        self.__salary = salary
        
    def get_employee_id(self):
        return int(self.__empid)
    
    def give_raise(self):
        if .02 * self.__salary > 2000:
            self.__salary = round(1.02 * self.__salary)
        else:
            self.__salary += 2000
#
# --------------------------------------------------------------------------------------------------------------

#infile = open("c:\\Projects\Continuing Education\2021\ITI456\Day_4\day4_employee_info.csv")

#
#Store employee information in a list of Employee classes

infile = open("day4_employee_info.csv")
emp_list = []

for info in infile:
#    print(info)
    temp = info.rsplit(",")
    emp_list.append(Employee(int(temp[0]), temp[1], temp[2], int(temp[3]), int(temp[4])))


#Create menu for user to select function(s)

print("\nMENU:\n")
print("\n1.) List Employees")
print("\n2.) Add Employee")
print("\n3.) Delete Employee")
print("\n4.) Update Employee Information")
print("\n5.) Give Raise")
print("\n6.) Quit")

choice = 1

while choice:
    choice = int(input("\nPlease enter menu option: "))
    
    if choice == 1:
        print ("\n list employees")
        e1 = 0    
        for emp in emp_list:
            print (emp_list[e1].print_emp_info())
            e1 +=1
#
# Add employee

    elif choice == 2:
        
        match = True
        while match == True:
            t_empid = int(input("Enter Employee ID:  "))
#
# Check if new employee ID is already in use
            e1 = 0    
            for emp in emp_list:
#  debugging              print ("\n in loop")
#  debugging              print(f"{(emp_list[e1].get_employee_id())}")
                if t_empid == (emp_list[e1].get_employee_id()):
                    print(f"\nEmployee ID {t_empid} is already in use. Please enter another ID.")
                    match = True
                    break
# Increment the counter (e1). If the new employee id does not match an existing id, set match to "False" so program will
# exit while loop. Otherwise, is "True" and the new employee id matches an existing id, the loop will start again to get 
# another new employee id.
                    
                else:
                    e1 +=1
                    if e1 == len(emp_list)-1:
                        match = False
                      
        t_name = input("Enter Employee Name: ")
        t_address = input("Enter Employee Address: ")
        t_date = int(input("Enter Year of employment: "))
        t_salary = int(input("Enter salary: "))
        emp_list.append(Employee(t_empid, t_name, t_address, t_date, t_salary))
#
# Delete employee

    elif choice == 3:
#       print ("\ndelete Employee")
        t_empid = int(input("Enter Employee ID of Employee to delete: "))
    
# Delete the list entry for the class that was just deleted
        emp_list.pop(t_empid-1)
        
#
# Update Employee information
        
    elif choice == 4:
     
        match = False
        while match == False:
            t_empid = int(input("Enter Employee ID of employee for update:  "))
#
# Check if new employee ID is already in use
            e1 = 0 
#           match = False
            for emp in emp_list:
#  debugging              print ("\n in loop")
# debugging               print(f"{(emp_list[e1].get_employee_id())}")
                if t_empid == (emp_list[e1].get_employee_id()):
#                    print(f"\nEmployee ID {t_empid} is already in use. Please enter another ID.")
                    match = True
                    break
# Increment the counter (e1). If the employee id does not match an existing id, set match to "False" so program will
# exit while loop. Otherwise, is "True" and the new employee id matches an existing id, the loop will start again to get 
# another new employee id.
                    
                else:
                    match = False
                    e1 +=1
                    if e1 == len(emp_list)-1:
                        print(f"\nEmployee ID {t_empid} does not exist. Please enter another ID.")
#                        match = False
 
        
        print(f"\nEmployee Update Menu for Employee {t_empid}:")
        print("\n1.) Update Name")
        print("\n2.) Update Address")
        print("\n3.) Update Salary")
        print("\n4.) Finished with updates")
        u_choice = 1
        while u_choice:
            u_choice = int(input("\nPlease enter menu option: "))
            if u_choice == 1:
                t_name = input("Enter New Employee Name: ")
                emp_list[t_empid-1].update_employee_name(t_name)
            elif u_choice == 2:
                t_address = input("Enter New Employee Address: ")
                emp_list[t_empid-1].update_employee_address(t_address)
            elif u_choice == 3:
                t_salary = input("Enter New Employee Salary: ")
                emp_list[t_empid-1].update_employee_salary(t_salary)
            elif u_choice == 4:
                break
            else:
                print ("\nInvalid option. Please enter 1, 2, 3, or 4")

#
# Give the greater of $2,000 or 2% raise to all employees
                
    elif choice == 5:
        e1 = 0    
        for emp in emp_list:
            emp_list[e1].give_raise()
            e1 +=1
                       
    elif choice == 6:
        break
        
    else:
        print ("\nInvalid option. Please enter 1, 2, 3, 4, 5 or 6")
        

# Create file for output. Overwrite if file exists.
#
fileout = open("employee_data_final.csv","w")
i1 = 0
for emp in emp_list:
    fileout.write(f"{emp_list[i1].print_emp_info_brief()}\n")
    i1+=1
    
fileout.close()
print ("\nPrograms ends")
```
